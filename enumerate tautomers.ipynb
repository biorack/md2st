```python
%matplotlib notebook
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
# load molvs library
from molvs import tautomer
from molvs import standardize
from molvs import standardize_smiles
```

```python
mol = Chem.MolFromSmiles('N[C@@H](Cc1c[nH]cn1)C(=O)O')
mol
```

```python
enum=tautomer.TautomerEnumerator()
tautomers=enum.enumerate(mol)
len(tautomers)
```

```python
img=Draw.MolsToGridImage(tautomers,molsPerRow=6,subImgSize=(200,200))
img
```

```python
canon = tautomer.TautomerCanonicalizer()
canon.canonicalize(mol)
```
